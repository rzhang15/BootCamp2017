```python
import numpy as np
from matplotlib import pyplot as plt
import math

def row_var(n):
    arr = np.random.randn(n,n)
    row_means = arr.mean(axis=1)
    var = np.var(row_means)
    return var

def problem_1():
    n = np.arange(100,1001,100)
    vfunc = np.vectorize(row_var)
    vars_n = vfunc(n)
    plt.plot(n, vars_n)
    plt.show()

problem_1()
```

```python
def problem_2():
    x = np.linspace(-2*np.pi, 2*np.pi, 100)
    sin = np.sin(x)
    cos = np.cos(x)
    plt.plot(x, sin)
    plt.show()
    plt.plot(x, cos)
    plt.show()

problem_2()
```

```python
def problem_3():
    x1 = np.arange(-3,1,0.0001)
    x2 = np.arange(1.0000001,7,0.0001)
    func = lambda x: 1/(x-1)
    plt.plot(x1, func(x1), 'm--', linewidth=5)
    plt.plot(x2, func(x2), 'm--', linewidth=5)
    plt.ylim(-6,6)
    plt.xlim(-2,6)
    plt.show()

problem_3()
```

```python
def problem_4():
    x = np.linspace(0,2*np.pi,100)

    fig, axes = plt.subplots(nrows=2, ncols=2)
    plt.subplot(221)
    plt.plot(x,np.sin(x),'g-')
    plt.axis([0, 2*np.pi, -2, 2])
    plt.title('Sin(x)')

    plt.subplot(222)
    plt.plot(x,np.sin(2*x),'r--')
    plt.axis([0, 2*np.pi, -2, 2])
    plt.title('Sin(2x)')

    plt.subplot(223)
    plt.plot(x,2*np.sin(x),'b--')
    plt.axis([0, 2*np.pi, -2, 2])
    plt.title('2sin(x)')

    plt.subplot(224)
    plt.plot(x,2*np.sin(2*x),'m:')
    plt.axis([0, 2*np.pi, -2, 2])
    plt.title('2sin(2x)')

    fig.subplots_adjust(hspace=.5)
    plt.suptitle('Different sine graphs')
    plt.show()

problem_4()
```

```python
def problem_5():
    fars = np.load("/Users/rubyzhang/Desktop/UChicago/OSML/BootCamp2017/Computation/Wk2_VisPandas/FARS.npy")
    long = np.array(fars[:,1])
    lat = np.array(fars[:,2])
    time = np.array(fars[:,0])

    fig, axes = plt.subplots(nrows=1, ncols=2)
    plt.subplot(121)
    plt.plot(long,lat,'ko',markersize=1)
    plt.axis('equal')
    plt.ylabel('Latitude')
    plt.xlabel('Longitude')

    plt.subplot(122)
    plt.hist(time,bins=np.arange(0,25))
    plt.xlabel('Hour of Day')
    fig.subplots_adjust(wspace=.5)

    plt.show()

problem_5()
```

```python
def problem_6():
    x = np.linspace(-2*np.pi, 2*np.pi, 100)
    y = x.copy()
    X, Y = np.meshgrid(x,y)
    Z = np.sin(X)*np.sin(Y)/(X*Y)

    fig, axes = plt.subplots(nrows=1, ncols=2)
    plt.subplot(121)
    plt.pcolormesh(X,Y,Z, cmap = "viridis")
    plt.colorbar()
    plt.xlim(-2*np.pi, 2*np.pi)
    plt.ylim(-2*np.pi, 2*np.pi)

    plt.subplot(122)
    plt.contour(X,Y,Z,15,cmap="magma")
    plt.colorbar()
    plt.xlim(-2*np.pi, 2*np.pi)
    plt.ylim(-2*np.pi, 2*np.pi)

    fig.subplots_adjust(hspace=.5)
    plt.show()

problem_6()
```
