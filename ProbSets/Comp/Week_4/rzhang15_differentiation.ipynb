```python
import sympy as sy
from sympy.utilities.lambdify import lambdify
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
def func(x):
    return ((math.sin(x)+1)**(math.sin(math.cos(x))))

from sympy.utilities.lambdify import lambdify
f_prime = lambdify(x, sy.diff((sy.sin(x) + 1)**(sy.sin(sy.cos(x)))),'numpy') # returns a numpy-ready function

f = np.vectorize(func)

x_1 = np.linspace(-np.pi,np.pi,100)
ax = plt.gca()
plt.plot(x_1,f(x_1),label='Original Function')
plt.plot(x_1,f_prime(x_1),label='Derivative Function')  
plt.legend(loc='best')
ax.spines["bottom"].set_position("zero")
plt.show()
```
